1.IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2.LOAD DATASET

```python
df=pd.read_csv('amazon-reviews.csv')
df.head()
```

3.REMOVING COLUMNS

```python
df = df[['Summary', 'Score']]
df.head()
```

4.REMOVING DUPLICATES

```python
df = df.drop_duplicates(subset=['Summary', 'Score'], keep='first')
df.head()
```

5.CHECKING MISSING VALUES

```python
df[['Summary', 'Score']].isnull().sum()
```

6.FILLING THE DATA

```python
df['Summary'] = df['Summary'].fillna(df['Summary'].mode()[0])
```

7.HANDLING THE DATA AGAIN

```python
# Check for any remaining missing values in 'Summary'
df['Summary'].isnull().sum()
```

```python
df[['Score', 'Summary']].isnull().sum()
```

```python
df.info()
```

8.DATA VIZULIZATION

8.1 RATING DISTRUBTUION SCORE

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['Score']  # only numerical column

plt.figure(figsize=(10, 5))

for i, col in enumerate(cols):
    plt.subplot(1, 1, i+1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

14.LOADING CLEANED DATASET

```python
import pandas as pd

df = pd.read_csv("cleaned_dataset.csv")
df.head()   
print("CLEANED DATASET LOADED")
```

```python
df[['Score', 'Summary']].isnull().sum()
```

15.DATA PRE-PROCESSING

15.1 REMOVING EXACT DUPLICATES

```python
df = df.drop_duplicates(subset=['Summary', 'Score'])
print("Exact duplicates removed:", df.shape)
```

15.2 REMOVING CONFLICT VALUE

```python
# Remove conflicting reviews (same Summary, different Score)
df = df[df.groupby('Summary')['Score'].transform('nunique') == 1]

print("Conflicting reviews removed")
print(df.shape)
```

16. DATA VIZUALIZATION

16.1 SCORE DISTRIBUTION BAR CHART

```python
import matplotlib.pyplot as plt

score_counts = df['Score'].value_counts().sort_index()

plt.figure()
plt.bar(score_counts.index, score_counts.values)
plt.xlabel('Score')
plt.ylabel('Number of Reviews')
plt.title('Distribution of Review Scores')
plt.show()
```

17.BALANCED DATASET

```python
target_count = 5000

balanced_df = (
    df.groupby('Score', group_keys=False)
      .sample(n=target_count, random_state=42)
)

print(balanced_df['Score'].value_counts())
```

18.SCORE DISPLAY

```python
print(balanced_df['Score'].value_counts())
```

```python
balanced_df.info()
```

19.BALANCED DATASET RATING COUNT

```python
#rating count plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=balanced_df['Score'])
plt.title("Balanced Dataset  Rating Count Plot")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

20.RATING DISTRIBUTION PIE CHART

```python
balanced_df['Score'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6)
)
plt.title("Balanced Dataset  Rating Distribution (Pie Chart)")
plt.ylabel("")
plt.show()
```

21.SAVING BALANCED DATASET

```python
balanced_df.to_csv("balanced_dataset.csv", index=False)
print("balanced dataset created successfully!")
```

22..IMBALANCED DATASET

```python
N = 25000

ratios = {
    1: 0.10,
    2: 0.15,
    3: 0.25,
    4: 0.30,
    5: 0.20
}

dfs = []

for rating, pct in ratios.items():
    count = int(N * pct)    # samples to take for this rating
    subset = df[df['Score'] == rating].sample(n=count, random_state=42)
    dfs.append(subset)
df_imbalanced = pd.concat(dfs).sample(frac=1, random_state=42).reset_index(drop=True)

df_imbalanced.shape
df_imbalanced['Score'].value_counts()
```

23.IMBALANCED RATING DISTRIBUTION

```python
import matplotlib.pyplot as plt

score_counts = df_imbalanced['Score'].value_counts().sort_index()

plt.figure()
plt.bar(score_counts.index, score_counts.values)
plt.xlabel('Rating (Score)')
plt.ylabel('Number of Reviews')
plt.title('Imbalanced Rating Distribution')
plt.show()
```

24.IMBALANCED RATING DISTRIBUTION

```python
import matplotlib.pyplot as plt

# Count ratings
score_counts = df_imbalanced['Score'].value_counts().sort_index()

# Pie chart
plt.figure()
plt.pie(
    score_counts,
    labels=score_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Imbalanced Rating Distribution')
plt.show()
```

25.SAVING IMBALANCED DATASET

```python
df_imbalanced.to_csv("imbalanced_dataset.csv", index=False)
print(" Imbalanced dataset created successfully!")
```
